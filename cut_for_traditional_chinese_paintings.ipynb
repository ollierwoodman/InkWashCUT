```python
NUMBER_OF_TRAINING_PHOTOGRAPHS = 100
HUA_CATEGORY = 'wgy_guo_hua_shan_shui'

for category in [
    # '', #landscapes 
    '_(2)', #mountains
    # '_(3)', #desert
    '_(4)', #sea 
    '_(5)', #beach
    # '_(6)', #island
    '_(7)', #japan
    ]:

    for i in range(NUMBER_OF_TRAINING_PHOTOGRAPHS):
      number = "{:08d}".format(i)
      filename = number + category + ".jpg"
      print("placed " + filename)
      !mkdir -p /content/contrastive-unpaired-translation/datasets/"$HUA_CATEGORY"/trainA/
      !cp /content/landscape-pictures/"$filename" /content/contrastive-unpaired-translation/datasets/"$HUA_CATEGORY"/trainA/
```
